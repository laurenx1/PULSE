```python
import pandas as pd

# Define the path to your CSV file in Google Drive
input_file_path = '/content/drive/My Drive/metaU/pb_transformed.csv'
output_file_path = '/content/drive/My Drive/metaU/pb_randomized_transformed_REAL.csv'


```

```python
df = pd.read_csv(input_file_path)

# Shuffle the DataFrame rows
shuffled_df = df.sample(frac=1).reset_index(drop=True)

# Write the shuffled DataFrame to a new CSV file
shuffled_df.to_csv(output_file_path, index=False)

print(f'Shuffled CSV saved as {output_file_path}')
```
