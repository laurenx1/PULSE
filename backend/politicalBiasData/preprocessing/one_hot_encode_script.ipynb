```python
import pandas as pd

# Define the path to your CSV file in Google Drive
input_file_path = '/content/drive/My Drive/metaU/political_bias_labeled_dataset.csv'
output_file_path = '/content/drive/My Drive/metaU/pb_transformed.csv'

# Read the original CSV file
df = pd.read_csv(input_file_path)

# Initialize the new columns with False
df['left-leaning'] = False
df['right-leaning'] = False
df['neutral'] = False

# Update the new columns based on the score column
df.loc[df['score'] == 'left-leaning', 'left-leaning'] = True
df.loc[df['score'] == 'right-leaning', 'right-leaning'] = True
df.loc[df['score'] == 'neutral', 'neutral'] = True

# Drop the original score column
df = df.drop(columns=['score'])

# Write the transformed data to a new CSV file
df.to_csv(output_file_path, index=False)

print('Transformation complete. The new file is saved to:', output_file_path)
```
